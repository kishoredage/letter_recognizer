```python
from sklearn.datasets import load_wine
import pandas as pd
import numpy as np
```

```python
df = load_wine(as_frame=True).frame
```

```python
df.head()
```

```python
df['target'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler

le=LabelEncoder()
scaler=StandardScaler()
```

```python
x = scaler.fit_transform(df.drop('target',axis=1))
y = le.fit_transform(df['target'])
```

```python
x
```

```python
y
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
DT = DecisionTreeClassifier()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=45)
```

```python
DT = DT.fit(x_train,y_train)
```

```python
y_pred = DT.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import*
```

```python
accuracy_score(y_test,y_pred)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

# Calculate the confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Get class labels from the LabelEncoder
class_labels = le.classes_

# Plot the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_labels, yticklabels=class_labels)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix')
plt.show()
```

## Visualize Actual vs. Predicted Clusters

### Subtask:
Create two scatter plots side-by-side. One plot will show the PCA-transformed `x_test` points colored by their `y_test` (actual labels), and the other will show the same points colored by their `y_pred` (predicted labels). This will allow for a visual comparison of the model's performance in identifying the different groups. Ensure that the plots have appropriate titles, axis labels, and legends.


**Reasoning**:
I will create two side-by-side scatter plots. The first plot will show the PCA-transformed test data colored by the actual labels (`y_test`), and the second plot will show the same PCA-transformed test data colored by the predicted labels (`y_pred`). This will allow for a visual comparison of actual versus predicted clusters, as specified in the subtask.



```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(14, 6))

# Plot 1: Actual Clusters
plt.subplot(1, 2, 1)
sns.scatterplot(x=x_test_pca[:, 0], y=x_test_pca[:, 1], hue=y_test, palette='viridis', legend='full')
plt.title('Actual Clusters (PCA)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')

# Plot 2: Predicted Clusters
plt.subplot(1, 2, 2)
sns.scatterplot(x=x_test_pca[:, 0], y=x_test_pca[:, 1], hue=y_pred, palette='viridis', legend='full')
plt.title('Predicted Clusters (PCA)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')

plt.tight_layout()
plt.show()
```

Visual Insights from PCA Scatter Plots:

Comparing the 'Actual Clusters (PCA)' and 'Predicted Clusters (PCA)' scatter plots reveals how well the Decision Tree Classifier is able to separate the different wine classes in a 2-dimensional PCA space.

- **Overall Cluster Separation**: Both plots generally show distinct clusters, indicating that the principal components effectively capture the variance necessary to differentiate the wine types.
- **Agreement between Actual and Predicted**: The predicted clusters largely align with the actual clusters, suggesting that the model has learned to classify most samples correctly. The spatial distribution and relative positions of the clusters appear similar in both plots.
- **Misclassifications**: Upon closer inspection, some points in the 'Predicted Clusters' plot might be colored differently than in the 'Actual Clusters' plot, particularly at the boundaries where clusters are less distinctly separated. These represent the misclassified samples. For example, if a point that is truly class 0 is predicted as class 1, its color will differ between the two plots.
- **Boundary Clarity**: The plots can also give an intuitive sense of the decision boundaries. Where the actual clusters are tightly grouped and well-separated, the predicted clusters tend to reflect this accuracy. Where actual clusters overlap, misclassifications are more likely to occur, and the predicted cluster boundaries might be less clean.

In summary, the side-by-side visualization confirms that the Decision Tree Classifier performs well in separating the wine classes, as indicated by the high accuracy score (0.916). The few discrepancies observed visually correspond to the misclassifications captured by the confusion matrix (e.g., 3 instances of class 2 being misclassified as class 1).
