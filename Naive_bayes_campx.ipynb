```python
# Calculate unnormalized posterior probabilities
unnormalized_p_yes = pyes * py
unnormalized_p_no = pno * pno

# Normalize the posterior probabilities
total_unnormalized_prob = unnormalized_p_yes + unnormalized_p_no

posterior_p_yes = unnormalized_p_yes / total_unnormalized_prob
posterior_p_no = unnormalized_p_no / total_unnormalized_prob

print(f"Posterior Probability of Playing (Yes): {posterior_p_yes:.4f}")
print(f"Posterior Probability of Not Playing (No): {posterior_p_no:.4f}")

if posterior_p_yes > posterior_p_no:
    print("Decision: Play (Yes)")
else:
    print("Decision: Do Not Play (No)")
```
